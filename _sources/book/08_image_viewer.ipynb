# Rapid Image Viewer

This chapter covers the development of a [rapid image viewer (RIV)](https://kenneth-ekpetere.users.earthengine.app/view/image-viewer-v3) on the Google Earth Engine (GEE). The RIV is a web application that allows users to view images in a browser and is built using Javascript and GEE GUI widgets. It is designed to be to be easy to use with a wide range of image types available on GEE.

Ken will be the primary author of the RIV.
