# Introduction

This chapter introduces the general concept of flood inundation mapping. It gives a review of the flood mapping process and the different components of the process. It also provides a brief overview of the different types of flood mapping methods and the different types of mapping tools used in the process.

Flooding is one of the most disruptive natural disasters that human societies face with global damages amounting to an estimated $651 billion (USD) and affecting 1.6 billion people between 2000 and 2019 alone according to the 2021 United Nations Office for Disaster Risk Reduction (UNDRR) annual report. Floods and their impacts are increasing in severity, duration, and frequency, mainly driven by population and economic growth in flood-prone areas, and by climate change. On the other hand, inundation of floodplains and wetlands plays a critical role in the Earth systems controlling the global dynamics of water, nutrient, and energy delivery to the coastal ocean and providing thermal, biogeochemical, and water vapor fluxes to the atmosphere at a global scale.

Mapping the extent and dynamics of surface water flooding is critical both to assessing flooding impacts and answering science questions. Three broad classes of information are needed by risk managers and scientist (Bates, 2022):
* Mapping areas at risk of flooding. A prediction of the area exposed to a flood event of a given probability, either now or in the future. This prediction can then be used for tasks such as land use zonation, exposure mapping, and loss analysis.
* Near-real-time inundation forecasts. To predict inundation patterns that are likely to occur in the immediate future as a result of some forecast high-water-level event.
* Floodplain inundation hindcasts. To map or simulate floodplain inundation that occurred in the past, either for specific events or for multiyear time series, to answer science questions and validate models.

## Low-complexity Flood Inundation Models for Near-real-time and Floodplain Inundation Mapping

* Why low-complexity FIMs are important

Low-complexity flood inundation models (FIMs) are a class of models that are computationally efficient and can be used to simulate flood inundation at a large scale. These models are used for near-real-time flood inundation mapping and floodplain inundation mapping. They are used to predict the extent of flooding and the depth of flooding in a region.
