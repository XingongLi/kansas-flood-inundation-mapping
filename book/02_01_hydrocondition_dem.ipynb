# Hydro-Condition DEM

The FLDPLN model requires a hydro-conditioned digital elevation model (DEM) as input. The hydro-conditioning process includes identifying depressions and filling shallow depressions, hydro-conditioning the DEM (dam and road culvert), adding flood defenses to the DEM, and calculating flow direction and flow accumulation. The hydro-conditioned DEM is then used to create stream networks and stream segments.

* Download DEM (typically in TIF format)
* Calculate hydrological derivatives from the DEM
    * Identify depressions and fill shallow depressions
    * Hydro-conditioning the DEM (dam and road culvert)
    * Add flood defenses to DEM
    * Calculate flow direction and flow accumulation
* Create BIL files for use in building FLDPLN library

### Add Breach Lines to Sinks and Lakes

While water can flow through culverts in reality, it is typically blocked by the DEM where the flow path is not represented and creates false sinks. In addition to culverts, lake/reservoir spillways may also not be represented on the DEM. We need to add breach lines on the DEM to represent how water actually moves/flows on the surface. 

### Mask Out Large Waterbody

After draining water from sinks and lakes/reservoirs, we want to mask out (i.e., set pixels to NODATA) large waterbodies as they are typically artificially controlled by USACE. Their inundation is caused by rising water level and can be simply modeled by a bathtub approach. Based on Kansas's experience, for most lakes/reservoirs, using their 75 percentile of elevation inundation extent is good choice to create the mask. Areas within those masks will not be modelled by the FLDPLN model.

### Add Flood Defense Structures

There generally 4 kind of flood defense structures that may need to be added onto the DEM. Most of them come from the [National Levee Database](https://levees.sec.usace.army.mil/) maintained by the USACE. Those structures include:

* flood gate, which can either contain water within the floodplain or let water flow out of the floodplain
* flood closure structure which is designed to contain the water within the floodplain
* flood wall, entire artificial wall to contain water within the floodplain
* levee, additional structure on top of some dirt/mud foundation

Jude developed a MATLAB script which turns levee 3D polylines (i.e., vertex has a Z value) into raster lines and added to DEM.
